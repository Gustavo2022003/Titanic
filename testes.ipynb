# .sum() definição e usabilidade

```python
import pandas as pd

data = {"A": [1, 2, None, 4], "B": [None, 2, 3, 4]}
df = pd.DataFrame(data)
```

Cada item do dicionário representa uma coluna:

{"A": [1, 2, None, 4], "B": [None, 2, 3, 4]}

| A    | B    |
|------|------|
| 1    | None |
| 2    | 2    |
| None | 3    |
| 4    | 4    |

`display()` -> printa a tabela (dataframe)

```python
display(df)
```

Caso eu queira mostrar o dataframe inteiro:

```python
# Configurar pandas para exibir todas as linhas e colunas
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
titanic_data = pd.read_csv("./titanic.csv")
display(titanic_data)
```

```python
# Restaurar configurações padrão de exibição de linhas e colunas
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')

display(titanic_data)
```

```python
display(df)
```

```python
# Somando os valores de cada coluna (eixo 0, que é o padrão)
soma_colunas = df.sum()
display(soma_colunas)
```

```python
# Somando os valores de cada linha (eixo 1)
soma_linhas = df.sum(axis=1)
display(soma_linhas)
```
